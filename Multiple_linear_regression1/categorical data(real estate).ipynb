```python
import matplotlib.pyplot as plt
import seaborn as sb
sb.set()
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv('/Users/dennis_m_jose/Downloads/real_estate_price_size_year_view.csv')
df
```

```python
data = df.copy()
data
```

```python
#maping the values of view to 0 and 1
data['view'] = data['view'].map({'No sea view':0, 'Sea view':1})
data 
```

```python
#describe the data
data.describe()
```

```python
Y = data['price']
x1 = data[['size','year','view']]

x = sm.add_constant(x1)

regression = sm.OLS(Y,x).fit()
regression.summary()
```
