## Handling Missing Values

```python
df = pd.read_csv('Nyc_Weather.csv',parse_dates = ["EST"])
df
```

```python
#Rename the column header EST to Day
df = df.rename(columns={'EST' : 'Day'})
df
```

```python
df
```

```python
# first way of handling NaN values => fillna
new_df = df.fillna(0)
new_df

#all the NaN values are changed to 0 by fillna() function
```

```python
#selectively handle the NaN values 
new_df = df.fillna({
    'Events' : 'No event',
    'Temperature' : 0
})
new_df.head()
```

```python
#forward fill 
new_df = df.fillna(method = 'ffill')
new_df
```

```python
# 2nd method dropping all the rows with NaN values
new_df = df.dropna()
new_df
```

## Group By


```python
weather = {
    'day' : ['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/2/2017','1/4/2017','1/2/2017','1/2/2017','1/3/2017','1/4/2017'],
    'City' : ['New York','New York','New York','New York','Mumbai','Mumbai','paris','paris','paris','paris'],
    'Temperature' : [32,34,29,30,32,31,34,34,29,30],
    'WindSpeed' : [6,7,8,10,15,4,15,3,6,7],
    'Event' : ['Rain', 'Snow', 'rain','Wind','Snow','Spring','Cloudy','Cloudy','Cloudy','Cloudy']
}

weather_df = pd.DataFrame(weather)
weather_df
```

## Q1) Find the avg windspeed per city
## Q2) Find the Maximum temperature in each city

```python
# Groupy by 
group_city = weather_df.groupby('City')
group_city

#This statement is very similar to that of SQL GroupBy Syntax : SELECT * From table groupby City
```

```python
# here city iterates through each city names and city_df is the corresponding grouped dataframe of the city
for city, city_df in group_city:
  print(city)
  print(city_df)
```

```python
#to get the group of particular city
group_city.get_group('New York')
```

```python
#Now let's move on to solve the Q2.) Finding the maximum temperature of each City
group_city.max()

# from the table we can see the maximum temperature of different cities.
```

```python
#Q1. Avg windspeed of each cities
group_city.mean()
```

```python
#to get all the statistics of each cities
group_city.describe()
```

# Concatenating DataFrames

```python
cars_df = pd.DataFrame({
    'Brand' : ['BMW','Mercedes','Skoda'],
    'Model' : [2018,2020,2009],
    'Owners' : ['Dennis','Sahal','Yadu']
})
cars_df
```

```python
bike_df = pd.DataFrame({
    'Brand' : ['Royal','Honda','Suzuki'],
    'Model' : [2016,2009,2010],
    'Owners' : ['Abhinand','Abhay','Aravind']
})
bike_df
```

```python
#concatenation 2 dataFrames
new_frame = pd.concat([cars_df,bike_df])
new_frame

#to get index correctly we can use the argument ignore_index = True
new_frame = pd.concat([cars_df,bike_df], ignore_index = True)
new_frame
```

```python
new_frame = pd.concat([cars_df,bike_df],ignore_index = True)
new_frame
```

```python
## to add a new column to the dataset
new_col = pd.Series(['200k','100k','150k','50k','47k','70k'],name = "Price")

new_df = pd.concat([new_frame,new_col],axis = 1)
new_df
```

#Merging DataFrames

```python
cars_price = pd.DataFrame({
    'Brand' : ['BMW','Mercedes','Honda','Skoda'],
    'Price' : [2000000,3000000,1800000,1800000]
})

car_df = pd.merge(cars_df,cars_price)
car_df
```

```python
car_df = pd.merge(cars_df,cars_price, how = 'outer')
car_df
```

```python
car_df = pd.merge(cars_df, cars_price, how = 'left')
car_df
```
